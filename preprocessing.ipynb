```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Untitled form (Responses) - Form Responses 1.csv")
df
```

```python
df.drop(labels = 'Timestamp' , axis = 1 , inplace = True)
```

```python
df.rename(columns = {'Specify your Gender' : 'gender',
          'Which programme are you enrolled in?' : 'program',
          'How frequently do you go home in a year? ' : 'freq',
          "How much money, on average, do you spend on one-way travel (from campus to home)? (in ₹)" : 'cost',
          'What is the average travel time, one way, in hours (from campus to home)?' : 'time',
          "What's your Annual Family Income? " : 'income',
          'State' : 'state',
          "What is your preferred mode of transport if the vacation is less than 10 days?  " : 'mode<10',
          "What is your preferred mode of transport if your vacation is more than 10 days? " : 'mode>10'},
        inplace=True)
df 
```

```python
df.loc[df["program"] == "Undergraduate (UG)" , "program"] = "ug"
df.loc[df["program"] == "PostGraduate (PG) + PhD" , "program"] = "pg"
df.loc[df["freq"] == "3  -  4 times" , "freq"] = "3-4"
df.loc[df["freq"] == "More than 10 times" , "freq"] = ">10"
df.loc[df["freq"] == "less than 2 times" , "freq"] = "<2"
df.loc[df["freq"] == "5  - 10 times" , "freq"] = "5-10"
df.loc[df["freq"] == "0  -  2 times" , "freq"] = "<2"
df.loc[df["income"] == "More than 5 Lakhs" , "income"] = ">5"
df.loc[df["income"] == "Less than or equal 5 Lakhs" , "income"] = "<5"
df.loc[df["mode<10"] == "Flight" , "mode<10"] = "flight"
df.loc[df["mode<10"] == "Train" , "mode<10"] = "train"
df.loc[df["mode<10"] == "By Road  (Bus/Car)" , "mode<10"] = "road"
df.loc[df["mode>10"] == "Flight" , "mode>10"] = "flight"
df.loc[df["mode>10"] == "Train" , "mode>10"] = "train"
df.loc[df["mode>10"] == "By Road  (Bus/Car)" , "mode>10"] = "road"
df.loc[df["gender"] == "Male" , "gender"] = "male"
df.loc[df["gender"] == "Female" , "gender"] = "female"
df
```

```python
df.drop(df[df['cost'] < 50].index , inplace = True)
df.drop(df[df['time'] < 1].index , inplace = True)
df.drop(df[(df['mode<10'] == 'flight') & (df['mode>10'] == 'flight') & (df['cost'] < 2000) ].index , axis=0 , inplace = True)
df.drop(df[(df['mode<10'] == 'train') & (df['mode>10'] == 'train') & (df['cost'] > 5000) ].index, axis=0 , inplace = True)
df.drop(df[(df['mode<10'] == 'train') & (df['mode>10'] == 'train') & (df['cost'] < 500) ].index, axis=0 , inplace = True)
df
```

```python
dist = pd.read_csv("State Distance.csv")
df = pd.merge(df, dist, on='state')
df.rename(columns = {'Distance' : 'dist'}, inplace=True)
```
