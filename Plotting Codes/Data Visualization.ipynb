```python
import pandas as pd
import numpy as np
from scipy.stats import t
from scipy.stats import chi2
import matplotlib.pyplot as plt
```

```python
alpha = 0.05  
mean = 0      
std_dev = 1   

critical_value_right = t.ppf(1-alpha/2, df = 240, loc=mean, scale=std_dev)

x = np.linspace(-10, 10, 1000)

pdf_values = t.pdf(x, df= 240,loc=mean, scale=std_dev)

plt.plot(x, pdf_values, 'b-', label='PDF')

plt.fill_between(x, 0, pdf_values, where=(x >= critical_value_right), color='red', alpha=0.3, label='Rejection Region')
plt.fill_between(x, 0, pdf_values, where=(x <= -critical_value_right), color='red', alpha=0.3,)

plt.axvline(x=critical_value_right, color='black', linestyle='--')
plt.axvline(x=-critical_value_right, color='black', linestyle='--')
plt.axvline(x=6.22, color='black', linestyle='--')
plt.text(6.22, -0.04, f'{6.22}', color='black', fontsize=8, ha='center')

plt.title('t - Distribution with Rejection Region for α = {}'.format(alpha))
plt.xlabel('X-axis')
plt.ylabel('Density')
plt.grid()
plt.legend()
plt.savefig('Hypo1.png')
plt.show()
```

```python
alpha = 0.05  
mean = 0      
std_dev = 1   

critical_value_right = chi2.ppf(1-alpha, df = 2, loc=mean, scale=std_dev)

x = np.linspace(0, 15, 1000)

pdf_values = chi2.pdf(x, df= 2,loc=mean, scale=std_dev)

plt.plot(x, pdf_values, 'b-', label='PDF')

plt.fill_between(x, 0, pdf_values, where=(x >= critical_value_right), color='red', alpha=0.3, label='Rejection Region')

plt.axvline(x=critical_value_right, color='black', linestyle='--')
plt.axvline(x=10.88, color='black', linestyle='--')
plt.text(10.88, -0.04, f'{10.88}', color='black', fontsize=8, ha='center')

plt.title('Chi - Squared Distribution with Rejection Region for α = {}'.format(alpha))
plt.xlabel('X-axis')
plt.ylabel('Density')
plt.grid()
plt.legend()
plt.savefig('Hypo2.png')
plt.show()
```
