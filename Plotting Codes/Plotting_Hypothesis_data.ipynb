```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('../data.csv')

distance = 750

nearBy = df[df['dist'] <= distance ]
farAway= df[df['dist'] > distance ]

print( nearBy.shape )
print( farAway.shape )

parts = np.zeros((4, 2), dtype=int)

parts[0, 0] = nearBy[nearBy['freq'] == '<2'].shape[0]
parts[0, 1] = farAway[farAway['freq'] == '<2'].shape[0]

parts[1, 0] = nearBy[nearBy['freq'] == '3-4'].shape[0]
parts[1, 1] = farAway[farAway['freq'] == '3-4'].shape[0]

parts[2, 0] = nearBy[nearBy['freq'] == '5-10'].shape[0]
parts[2, 1] = farAway[farAway['freq'] == '5-10'].shape[0]

parts[3, 0] = nearBy[nearBy['freq'] == '>10'].shape[0]
parts[3, 1] = farAway[farAway['freq'] == '>10'].shape[0]

plt.figure(figsize=(9, 6)) 

X = ['<= 750', '> 750']
bar_width = 0.35

plt.bar(X, parts[0], color='#bbd3da', width=bar_width, label="<2 times")
plt.bar(X, parts[1], bottom=parts[0], color='#519cab', width=bar_width, label="3-4 times")
plt.bar(X, parts[2], bottom=parts[0] + parts[1], color='#e3b451', width=bar_width, label="5-10 times")
plt.bar(X, parts[3], bottom=parts[0] + parts[1] + parts[2], color='#2d5157', width=bar_width, label=">10 times")

plt.xlabel("Distance from Hyderabad (in Kms)")
plt.ylabel("Number of Students")
plt.title("Distribution of people's visiting times relative to their distance")
plt.legend()

plt.tight_layout()
plt.savefig("../All Plots/Freq vs Dist", facecolor='w', bbox_inches="tight",
            pad_inches=0.3, transparent=True)
```
